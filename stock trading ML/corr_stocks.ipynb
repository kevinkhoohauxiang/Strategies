```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the URL of the Wikipedia page containing the S&P 500 list
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

# Fetch the tables from the Wikipedia page
tables = pd.read_html(url)

# Assuming that the first table on the page contains the S&P 500 list
sp500_df = tables[0]

# Extract the relevant columns (Symbol and Security)
sp500_companies = sp500_df['Symbol'].to_list()
print(sp500_companies)
```

```python
closing_prices_df = pd.DataFrame()

num_months = 12
end_date = datetime.now()
start_date = end_date - relativedelta(months=num_months)

# Loop through each ticker symbol and fetch the closing prices
for ticker in sp500_companies:
    try:
        # Download historical stock data for the specified date range
        stock_data = yf.download(ticker, start=start_date, end=end_date)
        
        # Extract the closing prices and store in the DataFrame
        closing_prices = stock_data['Close'].rename(ticker)
        closing_prices_df = pd.concat([closing_prices_df, closing_prices], axis=1)
    except Exception as e:
        print(f"Error fetching data for {ticker}: {str(e)}")

closing_prices_df.head()
```

```python
# Compute the correlation matrix
correlation_matrix = closing_prices_df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(50, 50))

sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=.5)

# Set the plot title
plt.title('Correlation Matrix of S&P 500 Company Closing Prices')

# Display the plot
plt.show()
```

```python
ticker = 'AAPL'

# Compute the correlation between AAPL and all other stocks
correlation_with_aapl = closing_prices_df.corr()[ticker].drop(ticker)

# Sort the correlations in descending order
sorted_correlation = correlation_with_aapl.sort_values(ascending=False)

# Get the top N stocks with the highest correlations (e.g., top 10)
top_N_stocks = sorted_correlation.head(20)

# Display the top N stocks
print("Top Stocks with Highest Correlation to "+ ticker)
print(top_N_stocks)
```

```python
# Now, we will add these stocks to the df, as the price movements of these stocks might be related to the movement of AAPL stock

num_months = 12
end_date = datetime.now()
start_date = end_date - relativedelta(months=num_months)

df = yf.download(ticker, start=start_date, end=end_date)
df = pd.concat([df,closing_prices_df[top_N_stocks.index.to_list()]], axis = 1)
df
```

```python
world_indices = [
    "^IXIC",  # NASDAQ Composite (U.S.)
    "^GSPC",  # S&P 500 (U.S.)
    "^DJI",   # Dow Jones Industrial Average (U.S.)
    "^N225",  # Nikkei 225 (Japan)
    "^FTSE",  # FTSE 100 (UK)
    "^GDAXI", # DAX (Germany)
    "^FCHI",  # CAC 40 (France)
    "^HSI",   # Hang Seng Index (Hong Kong)
    "^BSESN", # BSE SENSEX (India)
    "^AXJO",  # S&P/ASX 200 (Australia)
    "^MERV",  # Merval (Argentina)
    "^BVSP",  # IBOVESPA (Brazil)
    "^MXX",   # IPC (Mexico)
    "^STI"   # Straits Times Index (Singapore)
]

index_df = pd.DataFrame()

num_months = 12
end_date = datetime.now()
start_date = end_date - relativedelta(months=num_months)

# Loop through each ticker symbol and fetch the closing prices
for ticker in world_indices:
    try:
        # Download historical stock data for the specified date range
        stock_data = yf.download(ticker, start=start_date, end=end_date)
        
        # Extract the closing prices and store in the DataFrame
        closing_prices = stock_data['Close'].rename(ticker)
        index_df = pd.concat([index_df, closing_prices], axis=1)
    except Exception as e:
        print(f"Error fetching data for {ticker}: {str(e)}")

index_df = index_df[:len(index_df['^GSPC'].dropna())]
index_df
```
